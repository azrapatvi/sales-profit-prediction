```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ine-rmotr-curriculum/FreeCodeCamp-Pandas-Real-Life-Example/refs/heads/master/data/sales_data.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns=df.columns.str.lower()
```

```python
df['date']=pd.to_datetime(df['date'])
df.info()
```

```python
X=df.drop(['date','day','year','age_group','profit','cost','revenue'],axis=1)
y=df['profit']
```

```python
X.head(2)
```

```python
cat_cols=[i for i in X.columns if df[i].dtypes=='O']
num_cols=[i for i in X.columns if df[i].dtypes!='O']

cat_cols,num_cols
```

```python
df.head(2)
```

```python
#encoding

# 0->male
# 1->female

X['customer_gender']=X['customer_gender'].map({'M':0,'F':1})
```

```python
X.head(2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder

scaler=StandardScaler()
ohe=OneHotEncoder(drop='first',handle_unknown='ignore')

preprocessor=ColumnTransformer(
    [
        ("StandardScaler",scaler,num_cols),
        ("OneHotEncoder",ohe,cat_cols)
    ]
)
```

##### train test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.28,random_state=42)
```

```python
X_train_scaled=preprocessor.fit_transform(X_train)
X_test_scaled=preprocessor.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled.shape,X_test_scaled.shape
```

##### model training and selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor


```

```python
models={
    "LinearRegression":LinearRegression(),
    
    "DecisionTreeRegressor":DecisionTreeRegressor(random_state=42),
    "GradientBoostingRegressor":GradientBoostingRegressor(random_state=42),
    "AdaBoostRegressor":AdaBoostRegressor(random_state=42)
}
```

```python
models.items()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
for name,model in models.items():
    print(f"{name} training")
    print()

    model.fit(X_train_scaled,y_train)

    y_train_pred=model.predict(X_train_scaled)

    y_test_pred=model.predict(X_test_scaled)

    
    print("🔹 TRAINING PERFORMANCE")
    print("MAE:", mean_absolute_error(y_train, y_train_pred))
    print("MSE:", mean_squared_error(y_train, y_train_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
    print("R2 Score:", r2_score(y_train, y_train_pred))

    print()

    print("🔹 TESTING PERFORMANCE")
    print("MAE:", mean_absolute_error(y_test, y_test_pred))
    print("MSE:", mean_squared_error(y_test, y_test_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
    print("R2 Score:", r2_score(y_test, y_test_pred))
    print("\n" + "="*40 + "\n")
    print()
    
```

##### hyperparameter tuning

```python
gb=GradientBoostingRegressor()
```

```python
gb
```

```python

GradientBoosting_params={
        "n_estimators": [50, 100,120,20],
        "learning_rate": [0.01, 0.1],
        "max_depth": [3, 5,4,2],
        "subsample": [0.8, 1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(gb,param_grid=GradientBoosting_params,cv=5,n_jobs=-1,scoring='r2')

grid.fit(X_train_scaled, y_train)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
gb=GradientBoostingRegressor(learning_rate=0.1,max_depth=5,n_estimators=120,subsample=1.0)

gb.fit(X_train_scaled,y_train)

y_pred=gb.predict(X_test_scaled)

print("🔹 TESTING PERFORMANCE")
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score:", r2_score(y_test, y_pred))
print("\n" + "="*40 + "\n")
print()
```
